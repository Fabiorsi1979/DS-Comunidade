# SQL Operadores de Comparação

##### - Verificar sempre se os tipos de dados são iguais, por exemplo não é possível comparar 'texto/categoria' com 10 (numeros).
##### - As comparações podem ser feitas tanto nas cláusulas WHERE ou SELECT sendo essa segunda mais avançada.

### - Igualdade ( = )
### - Desigualdade ( <> )
### - Maio ( > ) ou Menor ( < )
### - Maior igual ( >= ) ou Menor igual ( <= )

```python
path = '/content/drive/MyDrive/db_olist.sqlite'
```

```python
import sqlite3
import pandas as pd
```

```python
def sql_compara (*args):
  conexao = sqlite3.connect(path)
  cursor = conexao.cursor()

  cursor.execute(consulta)
  dados = cursor.fetchall()

  cursor.close
  conexao.close

  consulta_sql = pd.DataFrame(dados, columns=colunas)
  return consulta_sql
```

#### 1. Qual o número de clientes únicos do estado de São Paulo?

```python
consulta = "SELECT COUNT(c.customer_id) FROM customer c WHERE c.customer_state ='SP'"
colunas  = ["Qtde_clientes_estado_SP"]

exercicio1 = sql_compara (consulta,colunas)
exercicio1
```

#### 2. Qual o número total de pedidos únicos feitos no dia 08 de Outubro de 2016.

```python
consulta = "SELECT COUNT(DISTINCT o.order_id) FROM orders o WHERE DATE(o.order_purchase_timestamp)='2016-10-08'"
colunas  = ["Pedidos_realizados"]

exercicio2 = sql_compara(consulta,colunas)
exercicio2
```

### 3. Qual o número total de pedidos únicos feitos a partir do dia 08 de Outubro de 2016

```python
consulta = "SELECT COUNT(DISTINCT o.order_id) FROM orders o WHERE DATE(o.order_purchase_timestamp)>='2016-10-08'"
colunas  = ["Pedidos_Realizados"]

exercicio3 = sql_compara(consulta,colunas)
exercicio3
```

### 4. Qual o número total de pedidos únicos feitos com a data limite de envio, a partir do dia 08 de Outubro de 2016 incluso

```python
consulta = "SELECT COUNT(DISTINCT o.order_id) FROM orders o WHERE o.order_delivered_carrier_date>='2016-10-08'"
colunas  = ["Pedidos realizados"]

exercicio4 = sql_compara(consulta,colunas)
exercicio4
```

### 5. Qual é o número total de pedidos únicos e o valor médio do frete para pedidos com valor abaixo de R$ 1.100?

```python
consulta = "SELECT COUNT(DISTINCT oi.order_id), AVG(oi.freight_value) FROM order_items oi WHERE oi.price <='1100'"
colunas  = ["Total_Pedidos","Media_frete"]

exercicio5 = sql_compara(consulta,colunas)
exercicio5
```

### 6. Qual é o número total de pedidos únicos, a data mínima e máxima de limite de envio, e os valores máximo, mínimo e médio do frete para pedidos com valor abaixo de R$1.100, agrupados por cada vendedor?


```python
consulta = "SELECT oi.seller_id, COUNT(DISTINCT oi.order_id), DATE(MIN(oi.shipping_limit_date)), DATE(MAX(oi.shipping_limit_date)), MAX(oi.freight_value), MIN(oi.freight_value), AVG(oi.freight_value) FROM order_items oi WHERE oi.price <='1100' GROUP BY oi.seller_id"
colunas  = ["Vendedor","Total_pedidos","Ultimo_embarque","Primeiro_embarque","Maior_frete","Menor_frete","Frete_medio"]

exercicio6 = sql_compara(consulta,colunas)
exercicio6
```
