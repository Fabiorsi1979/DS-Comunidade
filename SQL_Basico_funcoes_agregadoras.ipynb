```python
# Cria a função que será usada para executar as clausulas SQL no ambiente colab/ python
def consulta_sql_agg (*args):
    #consulta = input(str("Digite a clausula SQL a ser pesquisada"))
    #colunas = [input(str("Quais as colunas devem ser pesquisadas "))]
    conexao = sqlite3.connect(path)
    cursor = conexao.cursor()

    cursor.execute(consulta)
    resultado = cursor.fetchall()

    cursor.close
    conexao.close

    consulta_df = pd.DataFrame(resultado,columns=[colunas])
    return consulta_df
```

#### 1. Qual o número de clientes únicos do estado de Minas Gerais?

```python
consulta = "SELECT COUNT (c.customer_unique_id) AS Total_de_Clientes_MG FROM customer c WHERE c.customer_state='MG'"
colunas =['customer_unique_id']

exercicio1= consulta_sql_agg(consulta,colunas)
print("Total de Clientes do Estado de MG :")
exercicio1
```

#### 2. Qual a quantidade de cidades únicas dos vendedores do estado de Santa Catarina?


```python
consulta ="SELECT COUNT (DISTINCT ss.seller_city) AS Cidades_Vendedores_SC FROM sellers ss WHERE ss.seller_state='SC'"
colunas = ["seller_city"]

exercicio2 = consulta_sql_agg(consulta,colunas)
print("Total de cidades onde moram os vendedores de SC:")
exercicio2
```
