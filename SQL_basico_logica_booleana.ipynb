# SQL operadores de lógica Booleana

#### A lógica booleana é um sistema composto por operadores lógicos e binários:
##### ( AND ) Obrigatório que TODAs comparações sejam VERDADEIRAS.
##### ( OR ) Obrigatório que APENAS UMA das comparações sejam VERDADEIRAS.

##### obs:  É possível usar estas clausulas em SELECT e WHERE, sendo a primeira no modulo mais avançado.

```python
import sqlite3
import pandas as pd
```

```python
def sql_booleana (*args):
    conexao = sqlite3.connect (path)
    cursor = conexao.cursor()

    cursor.execute(consulta)
    dados = cursor.fetchall()

    cursor.close
    conexao.close

    consulta_sql = pd.DataFrame(data=dados,columns=colunas)
    return consulta_sql
```

### 1. Qual o número de clientes únicos nos estado de Minas Gerais ou Rio de Janeiro?


```python
consulta = "SELECT c.customer_state, COUNT (DISTINCT c.customer_id) FROM customer c WHERE c.customer_state ='MG' OR c.customer_state='RJ' GROUP BY c.customer_state"
colunas =["UF","Qtde_Clientes"]

exercicio1 = sql_booleana(consulta,colunas)
exercicio1
```

### 2. Qual a quantidade de cidades únicas dos vendedores no estado de São Paulo ou Rio de Janeiro com a latitude maior que -24.54 e longitude menor que -45.63?

```python
consulta = "SELECT sl.seller_state, COUNT(DISTINCT sl.seller_city) FROM sellers sl INNER JOIN geolocation ge ON (sl.seller_zip_code_prefix = ge.geolocation_zip_code_prefix) WHERE (sl.seller_state='SP' OR sl.seller_state='RJ') AND ge.geolocation_lat>-24.54 AND ge.geolocation_lng<-45.63 GROUP BY sl.seller_state"
colunas =["Estados","Total_cidades"]

exercicio2 = sql_booleana(consulta,colunas)
exercicio2
```

### 3. Qual o número total de pedidos únicos, o número total de produtos e o preço médio dos pedidos com o preço de frete maior que R$ 20 e a data limite de envio entre os dias 1 e 31 de Outubro de 2016? **OBS: Use a função DATE( ) para converter a data no formato timestamp (data e hora) para data.


```python
consulta = "SELECT COUNT(DISTINCT oi.order_id), COUNT(oi.product_id),AVG(oi.price) FROM order_items oi WHERE oi.freight_value>20 AND DATE(oi.shipping_limit_date >='2016-10-01' AND oi.shipping_limit_date <='2016-10-31')"
colunas =["Pedidos","Produtos","Frete_medio"]

exercicio3 = sql_booleana(consulta,colunas)
exercicio3
```

### 4. Mostre a quantidade total dos pedidos e o valor total do pagamento, para pagamentos entre 1 e 5 prestações ou um valor de pagamento acima de R$ 5000. Agrupe por quantidade de prestações.


```python
consulta = "SELECT op.payment_installments, COUNT(op.order_id), SUM(op.payment_value) FROM order_payments op WHERE ((op.payment_installments>=1 AND op.payment_installments<=5) OR op.payment_value>5000) GROUP BY op.payment_installments"
colunas =["Qtde_Prestacoes", "Total_pedidos","Total_valores"]

exercicio4 = sql_booleana(consulta,colunas)
exercicio4
```

### 5. Qual a quantidade de pedidos com o status em processamento ou cancelada acontecem com a data estimada de entrega maior que 01 de Janeiro de 2017 ou menor que 23 de Novembro de 2016?

```python
consulta = "SELECT COUNT (o.order_id) FROM orders o WHERE (o.order_status= 'processing' OR o.order_status= 'canceled') AND (o.order_estimated_delivery_date > '2017-01-01'AND o.order_estimated_delivery_date < '2016-11-23')"
colunas =["Total_pedidos_periodo"]

exercicio5 = sql_booleana(consulta,colunas)
print('No período de 01-01-2017 até 23-11-2016 temos:')
exercicio5
```

#### 6. Quantos produtos estão cadastrados nas categorias: perfumaria, brinquedos, esporte lazer, cama mesa e banho e móveis de escritório que possuem mais de 5 fotos, um peso maior que 5 g, um altura maior que 10 cm, uma largura maior que 20 cm?

```python
consulta = "SELECT p.product_category_name, COUNT(p.product_id) FROM products p WHERE p.product_category_name IN ('perfumaria','brinquedos','esporte_lazer','cama_mesa_banho','moveis_escritorio') AND p.product_photos_qty>5 AND p.product_weight_g>5 AND p.product_height_cm>10 AND p.product_length_cm>20 GROUP BY p.product_category_name"
colunas =["Categoria","Qtde_produtos"]

exercicio6 = sql_booleana(consulta,colunas)
exercicio6
```
