## Funções de Agrupamento - SQL com Phyton

#####  - As funções de agrupamentos unificam categorias para que seja possível funções agregadoras e insights
#####  - Utiliza-se a clausula GROUP BY para indicar as colunas que estão sendo agrupadas

```python
import pandas as pd
import sqlite3
```

```python
path = ("/content/drive/MyDrive/db_olist.sqlite")
```

```python
def sql_agrup (*args):
    conexao = sqlite3.connect(path)
    cursor = conexao.cursor()

    cursor.execute(consulta)
    dados = cursor.fetchall()

    cursor.close
    conexao.close

    consulta_sql = pd.DataFrame(dados, columns=[colunas])
    return consulta_sql
```

#### 1. Qual o número de clientes únicos de todos os estados?

```python
consulta = "SELECT c.customer_state, COUNT (DISTINCT c.customer_id) FROM customer c GROUP BY c.customer_state"
colunas = ["UF","Quant_clientes"]

exercicio1 = sql_agrup(consulta,colunas)
exercicio1
```

#### 2. Qual o número de cidades únicas de todos os estados?


```python
consulta = "SELECT geo.geolocation_state, COUNT (DISTINCT geo.geolocation_city) FROM geolocation geo GROUP BY geo.geolocation_state"
colunas =  ["UF","Qtd_cidades"]

exercicio2 = sql_agrup(consulta,colunas)
exercicio2
```
