# Exercícios

```python
def consulta_sql (consulta,colunas):
    conexao = sqlite3.connect(path)
    cursor = conexao.cursor()

    cursor.execute(consulta)
    dados = cursor.fetchall()

    cursor.close()
    conexao.close()

    consulta_df = pd.DataFrame(dados, columns=[colunas])
    return consulta_df
```

#### 1. Gere uma tabela com o id do cliente, a cidade e estado onde ele vive.

```python
consulta = "SELECT customer_id, customer_city, customer_state FROM customer "
colunas = ["codigo","Cidade","UF"]

exercicio1 = consulta_sql(consulta,colunas)
exercicio1.head()
```

####2. Gere uma tabela com o id do cliente e a cidade,**somente** dos clientes que vivem em Santa Catarina.

```python
consulta = "SELECT customer_id, customer_city FROM customer WHERE customer_state ='SC' ORDER BY customer_city"
colunas=["codigo","Cidade"]

exercicio2 = consulta_sql (consulta, colunas)
exercicio2.head()
```
