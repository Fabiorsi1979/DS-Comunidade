# LEFT JOIN

#### O LEFT JOIN é usado quando queremos tornar uma tabela como base para puxar das demais informações, quando não tem correspondente na outra tabela, ele mantem a base com informações nulas onde a tabela pesquisada não tem correspondente

```python
import sqlite3
import pandas as pd
```

```python
def sql_LEFT (*args):
  conexao = sqlite3.connect(path)
  cursor = conexao.cursor()

  cursor.execute(consulta)
  dados = cursor.fetchall()

  consulta_sql = pd.DataFrame(data=dados, columns=colunas)
  return consulta_sql
```

#### 1. Gerar uma tabela de dados com 20 linhas e contendo as seguintes colunas: 1) Id do pedido, 2) status do pedido, 3) id do produto, 4) categoria do produto, 5) avaliação do pedido, 6) valor do pagamento, 7) tipo do pagamento, 8) cidade do vendedor, 9) latitude e longitude da cidade do vendedor.

```python
consulta ="""
SELECT
     o.order_id,
     o.order_status,
     oi.product_id,
     p.product_category_name,
     ow.review_score,
     op.payment_value,
     op.payment_type,
     s.seller_city,
     g.geolocation_lat,
     g.geolocation_lng
FROM orders o LEFT JOIN order_items oi    ON ( oi.order_id = o.order_id )
              LEFT JOIN products p        ON ( p.product_id = oi.product_id )
              LEFT JOIN order_reviews ow  ON ( ow.order_id = o.order_id )
              LEFT JOIN order_payments op ON ( op.order_id = o.order_id )
              LEFT JOIN sellers s         ON ( s.seller_id = oi.seller_id )
              LEFT JOIN geolocation g     ON ( g.geolocation_zip_code_prefix = s.seller_zip_code_prefix )
LIMIT 20
"""
colunas =["Codigo Pedido","Status","Codigo Produto","Categoria","Avaliação Pedido","Valor Pagamento","Tipo Pagamento","Cidade Vendedor","Lat. Vendedor","Long. Vendedor"]

resultado = sql_LEFT (consulta,colunas)
resultado
```

#### 2. Quantos tipos de pagamentos foram usados pelo cliente para pagar o pedido 'e481f51cbdc54678b7cc49136f2d6af7’

```python
# e481f51cbdc54678b7cc49136f2d6af7
consulta ="""
SELECT
     op.payment_type,
     o.order_id
FROM orders o LEFT JOIN order_payments op ON ( op.order_id = o.order_id)
WHERE o.order_id = 'e481f51cbdc54678b7cc49136f2d6af7'
"""
colunas =["Tipo de Pagamento","Codigo Pedido"]

resultado = sql_LEFT (consulta,colunas)
resultado
```

#### 3. Quantos pedidos tem mais de 5 items?


```python
consulta ="""
SELECT
  o.order_id,
  COUNT(DISTINCT oi.product_id)
FROM orders o LEFT JOIN order_items oi ON ( oi.order_id = o.order_id )
GROUP BY o.order_id
HAVING COUNT(DISTINCT oi.product_id)>5
"""
colunas =["Pedidos","Qtde Items"]

resultado = sql_LEFT (consulta,colunas)
resultado
```

#### 4. Qual a cardinalidade entre a tabela Pedidos ( orders ) e Avaliações ( reviews )?


```python
consulta ="""
SELECT
  o.order_id,
  COUNT(ow.review_id)
FROM orders o LEFT JOIN order_reviews ow ON ( ow.order_id = o.order_id )
GROUP BY o.order_id
HAVING COUNT(ow.review_id)>1
"""
colunas =["Pedidos ID","Reviews ID"]

resultado = sql_LEFT (consulta,colunas)
resultado
```

 #### 5. Quantos pedidos (orders) não tem nenhuma avaliação (review) ?


```python
consulta ="""
SELECT
  o.order_id,
  COUNT(ow.review_id)
FROM orders o LEFT JOIN order_reviews ow ON ( ow.order_id = o.order_id )
GROUP BY o.order_id
HAVING COUNT(ow.review_id) IS NULL
"""
colunas =["Pedidos ID","Reviews ID"]

resultado = sql_LEFT (consulta,colunas)
resultado
```

#### 6. Quais são os top 10 vendedores com mais clientes?

```python
consulta ="""
SELECT
  s.seller_id,
  COUNT(c.customer_id) AS ccustomerid
FROM orders o LEFT JOIN order_items oi ON ( oi.order_id = o.order_id )
              LEFT JOIN sellers s      ON ( s.seller_id = oi.seller_id )
              LEFT JOIN customer c     ON ( c.customer_id = o.customer_id )
GROUP BY s.seller_id
ORDER BY ccustomerid DESC
LIMIT 10
"""
colunas =["Vendedores","Qtde Clientes"]

resultado = sql_LEFT (consulta,colunas)
resultado
```

#### 7. Quantos pedidos (orders) não possuem nenhum produto (products) ?

```python
consulta = """
SELECT
  COUNT( o.order_id )
FROM orders o LEFT JOIN order_items oi ON ( oi.order_id = o.order_id )
              LEFT JOIN products p     ON ( p.product_id = oi.product_id )
WHERE p.product_id IS NULL
"""

colunas =["Pedidos sem produtos",]

resultado = sql_LEFT (consulta,colunas)
resultado
```
